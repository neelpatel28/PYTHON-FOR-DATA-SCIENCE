```python
print("\n")
print("\tGame Start\nNow the task is to move all of them to right side of the river")
print('''rules:
1. The boat can carry at most two people
2. If cannibals num greater than
missionaries then the cannibals would eat the missionaries
3. The boat cannot cross the
river by itself with no people on boar''')

left_missionaries = 3
left_cannibals = 3
right_missionaries = 0
right_cannibals = 0

attempts = 0

def show_state(direction):
    print("\n")
    for _ in range(left_missionaries):
        print("M ", end="")
    for _ in range(left_cannibals):
        print("C ", end="")

    print(f"| {direction} | ", end="")

    for _ in range(right_missionaries):
        print("M ", end="")
    for _ in range(right_cannibals):
        print("C ", end="")
    print("\n")

print("\nM M M C C C | --- | \n")

try:
    while True:

        while True:
            print("Left side -> right side river travel")
            move_m = int(input("Enter number of Missionaries travel => "))
            move_c = int(input("Enter number of Cannibals travel => "))

            if move_m == 0 and move_c == 0:
                print("Empty travel not possible")
                print("Re-enter : ")

            elif (move_m + move_c <= 2 and
                  left_missionaries - move_m >= 0 and
                  left_cannibals - move_c >= 0):
                break
            else:
                print("Wrong input re-enter : ")

        left_missionaries -= move_m
        left_cannibals -= move_c
        right_missionaries += move_m
        right_cannibals += move_c

        attempts += 1
        show_state("-->")

        if ((left_cannibals == 3 and left_missionaries in (1, 2)) or
            (left_cannibals == 2 and left_missionaries == 1) or
            (right_cannibals == 3 and right_missionaries in (1, 2)) or
            (right_cannibals == 2 and right_missionaries == 1)):
            print("Cannibals eat missionaries:\nYou lost the game")
            break

        if right_missionaries + right_cannibals == 6:
            print("You won the game : \n\tCongrats")
            print("Total attempt")
            print(attempts)
            break

        while True:
            print("Right side -> Left side river travel")
            move_m = int(input("Enter number of Missionaries travel => "))
            move_c = int(input("Enter number of Cannibals travel => "))

            if move_m == 0 and move_c == 0:
                print("Empty travel not possible")
                print("Re-enter : ")

            elif (move_m + move_c <= 2 and
                  right_missionaries - move_m >= 0 and
                  right_cannibals - move_c >= 0):
                break
            else:
                print("Wrong input re-enter : ")

        left_missionaries += move_m
        left_cannibals += move_c
        right_missionaries -= move_m
        right_cannibals -= move_c

        attempts += 1
        show_state("<--")

        if ((left_cannibals == 3 and left_missionaries in (1, 2)) or
            (left_cannibals == 2 and left_missionaries == 1) or
            (right_cannibals == 3 and right_missionaries in (1, 2)) or
            (right_cannibals == 2 and right_missionaries == 1)):
            print("Cannibals eat missionaries:\nYou lost the game")
            break

except EOFError:
    print("\nInvalid input please retry !!")
```
